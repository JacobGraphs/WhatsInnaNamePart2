```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/new_data/final.csv')
```

```python
subset = ['title', 'duration', 'year', 'budget', 'is_Sport', 'is_Action', 'is_Adventure', 'is_Mystery', 'is_Comedy',
         'is_Thriller', 'is_History', 'is_Family', 'is_Western', 'is_War', 'is_Musical', 'is_Crime', 'is_Film-Noir',
         'is_Music', 'is_Romance', 'is_Drama', 'is_Animation', 'is_Horror', 'is_Biography', 'is_Sci-Fi',
         'is_Fantasy', 'profitable']
```

```python
df = df[subset]
```

```python
X_columns = ['duration', 'year', 'budget', 'is_Sport', 'is_Action', 'is_Adventure', 'is_Mystery', 'is_Comedy',
         'is_Thriller', 'is_History', 'is_Family', 'is_Western', 'is_War', 'is_Musical', 'is_Crime', 'is_Film-Noir',
         'is_Music', 'is_Romance', 'is_Drama', 'is_Animation', 'is_Horror', 'is_Biography', 'is_Sci-Fi',
         'is_Fantasy', 'nb_prediction']

X_cols = ['duration', 'budget', 'is_Sport', 'is_Action', 'is_Adventure', 'is_Mystery', 'is_Comedy',
         'is_Thriller', 'is_History', 'is_Family', 'is_Western', 'is_War', 'is_Musical', 'is_Crime', 'is_Film-Noir',
         'is_Music', 'is_Romance', 'is_Drama', 'is_Animation', 'is_Horror', 'is_Biography', 'is_Sci-Fi',
         'is_Fantasy']
X = ['duration', 'budget']
```

```python
train_data, test_data = train_test_split(df, test_size = .2, )
```

```python
counts = df[X].values
targets = df['profitable'].values
logreg = LogisticRegression()

logreg.fit(counts, targets)
```

```python
predictions = logreg.predict(counts)
df['lr_prediction'] = predictions
```

```python
for col, coef in zip(X_cols, logreg.coef_[0]):
    print(f'{col} {coef:0.6f}')
```
